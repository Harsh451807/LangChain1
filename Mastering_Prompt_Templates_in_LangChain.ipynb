```python
# Create multi-input prompt template
multi_input_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

def generate_multi_input_prompt(topic: str, audience: str, tone: str) -> str:
    """
    Generate customized prompts based on topic, audience, and tone

    Args:
        topic (str): Subject matter
        audience (str): Target audience
        tone (str): Communication style

    Returns:
        str: Formatted prompt
    """
    return multi_input_template.format(
        topic=topic,
        audience=audience,
        tone=tone
    )

# Test Cases
test_cases = [
    {"topic": "AI", "audience": "beginners", "tone": "friendly"},
    {"topic": "Python", "audience": "kids", "tone": "fun"},
    {"topic": "Deep Learning", "audience": "engineers", "tone": "technical"}
]

print("\n" + "=" * 60)
print("TASK 2: Multi-Input Prompt System")
print("=" * 60)
for i, test_case in enumerate(test_cases, 1):
    print(f"\n🔹 Test Case {i}:")
    print(f"   Input: {test_case}")
    print(f"   Output: {generate_multi_input_prompt(**test_case)}")
print("\n✅ Task 2 Completed Successfully!")
```

```python
# Create three distinct prompt templates
teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 questions about {topic}"
)

storytelling_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

class PromptVariationsEngine:
    """
    Engine to generate different types of prompts based on style
    """

    def __init__(self):
        self.templates = {
            "teaching": teaching_template,
            "interview": interview_template,
            "storytelling": storytelling_template
        }

    def generate(self, style: str, topic: str) -> str:
        """
        Generate prompt based on selected style

        Args:
            style (str): Type of prompt (teaching/interview/storytelling)
            topic (str): Subject matter

        Returns:
            str: Formatted prompt
        """
        if style not in self.templates:
            raise ValueError(f"Invalid style. Choose from: {list(self.templates.keys())}")

        return self.templates[style].format(topic=topic)

    def generate_all(self, topic: str) -> Dict[str, str]:
        """
        Generate all prompt variations for a given topic

        Args:
            topic (str): Subject matter

        Returns:
            dict: All prompt variations
        """
        return {
            style: template.format(topic=topic)
            for style, template in self.templates.items()
        }

# Initialize the engine
prompt_engine = PromptVariationsEngine()

# Test with Machine Learning
test_topic = "Machine Learning"

print("\n" + "=" * 60)
print("TASK 3: Prompt Variations Engine")
print("=" * 60)
print(f"\n📚 Topic: {test_topic}\n")

variations = prompt_engine.generate_all(test_topic)
for style, prompt in variations.items():
    print(f"🎯 {style.upper()} Style:")
    print(f"   {prompt}\n")

print("✅ Task 3 Completed Successfully!")
```

```python
from langchain_core.prompts import ChatPromptTemplate, SystemMessagePromptTemplate, HumanMessagePromptTemplate

class ChatPromptSystem:
    """
    Role-based chat prompt system using ChatPromptTemplate
    """

    def __init__(self):
        # Define system messages for different roles
        self.system_messages = {
            "teacher": "You are an experienced educator who explains concepts clearly and patiently. Your goal is to ensure students understand the fundamentals.",
            "interviewer": "You are a professional interviewer conducting a technical interview. Ask insightful questions to assess understanding.",
            "motivator": "You are an enthusiastic motivational coach. Inspire learners and make complex topics exciting and achievable."
        }

    def create_chat_prompt(self, role: str, topic: str) -> ChatPromptTemplate:
        """
        Create a chat prompt template for specific role and topic

        Args:
            role (str): User role (teacher/interviewer/motivator)
            topic (str): Subject matter

        Returns:
            ChatPromptTemplate: Configured chat prompt
        """
        if role not in self.system_messages:
            raise ValueError(f"Invalid role. Choose from: {list(self.system_messages.keys())}")

        # Create system message template
        system_template = SystemMessagePromptTemplate(
            prompt=PromptTemplate(
                input_variables=[],
                template=self.system_messages[role]
            )
        )

        # Create human message template
        human_template = HumanMessagePromptTemplate(
            prompt=PromptTemplate(
                input_variables=["topic"],
                template="Please help me understand {topic}"
            )
        )

        # Combine into chat prompt
        chat_prompt = ChatPromptTemplate.from_messages([
            system_template,
            human_template
        ])

        return chat_prompt

    def generate_chat_prompt(self, role: str, topic: str) -> List[Dict]:
        """
        Generate formatted chat messages

        Args:
            role (str): User role
            topic (str): Subject matter

        Returns:
            list: Formatted chat messages
        """
        chat_prompt = self.create_chat_prompt(role, topic)
        messages = chat_prompt.format_messages(topic=topic)

        return [
            {"role": msg.type, "content": msg.content}
            for msg in messages
        ]

# Initialize chat system
chat_system = ChatPromptSystem()

# Test cases
test_cases = [
    {"role": "teacher", "topic": "Neural Networks"},
    {"role": "interviewer", "topic": "Machine Learning Algorithms"},
    {"role": "motivator", "topic": "Deep Learning"}
]

print("\n" + "=" * 60)
print("TASK 4: ChatPromptTemplate System")
print("=" * 60)

for i, test_case in enumerate(test_cases, 1):
    print(f"\n{'=' * 60}")
    print(f"🎭 Test Case {i}: {test_case['role'].upper()} - {test_case['topic']}")
    print(f"{'=' * 60}")

    messages = chat_system.generate_chat_prompt(**test_case)

    for msg in messages:
        print(f"\n🔸 {msg['role'].upper()}:")
        print(f"   {msg['content']}")

print("\n✅ Task 4 Completed Successfully!")
```

```python
class InputValidator:
    """
    Validate inputs for prompt generation with strict rules
    """

    VALID_AUDIENCES = ["beginner", "intermediate", "expert"]
    VALID_TONES = ["formal", "casual", "fun"]

    # Default values
    DEFAULT_AUDIENCE = "beginner"
    DEFAULT_TONE = "casual"

    @classmethod
    def validate_inputs(cls, audience: str, tone: str, strict_mode: bool = False) -> Dict[str, str]:
        """
        Validate audience and tone inputs

        Args:
            audience (str): Target audience level
            tone (str): Communication tone
            strict_mode (bool): If True, raise errors; if False, use defaults

        Returns:
            dict: Validated inputs

        Raises:
            ValueError: If strict_mode=True and inputs are invalid
        """
        validated = {}
        errors = []

        # Validate audience
        if audience.lower() not in cls.VALID_AUDIENCES:
            error_msg = f"Invalid audience '{audience}'. Valid options: {cls.VALID_AUDIENCES}"
            if strict_mode:
                errors.append(error_msg)
            else:
                print(f"⚠️  {error_msg}")
                print(f"   Using default: '{cls.DEFAULT_AUDIENCE}'")
                validated["audience"] = cls.DEFAULT_AUDIENCE
        else:
            validated["audience"] = audience.lower()

        # Validate tone
        if tone.lower() not in cls.VALID_TONES:
            error_msg = f"Invalid tone '{tone}'. Valid options: {cls.VALID_TONES}"
            if strict_mode:
                errors.append(error_msg)
            else:
                print(f"⚠️  {error_msg}")
                print(f"   Using default: '{cls.DEFAULT_TONE}'")
                validated["tone"] = cls.DEFAULT_TONE
        else:
            validated["tone"] = tone.lower()

        # Raise errors if in strict mode
        if strict_mode and errors:
            raise ValueError("\n".join(errors))

        return validated

    @classmethod
    def get_valid_options(cls) -> Dict[str, List[str]]:
        """
        Get all valid input options

        Returns:
            dict: Valid audiences and tones
        """
        return {
            "audiences": cls.VALID_AUDIENCES,
            "tones": cls.VALID_TONES
        }

# Test the validation layer
print("\n" + "=" * 60)
print("TASK 5: Input Validation Layer")
print("=" * 60)

# Test Case 1: Valid inputs
print("\n🔹 Test Case 1: Valid Inputs")
result = InputValidator.validate_inputs("beginner", "formal")
print(f"✅ Validated: {result}")

# Test Case 2: Invalid inputs (default mode)
print("\n🔹 Test Case 2: Invalid Inputs (Default Mode)")
result = InputValidator.validate_inputs("novice", "friendly", strict_mode=False)
print(f"✅ Validated: {result}")

# Test Case 3: Invalid inputs (strict mode)
print("\n🔹 Test Case 3: Invalid Inputs (Strict Mode)")
try:
    result = InputValidator.validate_inputs("novice", "friendly", strict_mode=True)
except ValueError as e:
    print(f"❌ Error caught:\n   {e}")

# Test Case 4: Display valid options
print("\n🔹 Valid Options:")
valid_options = InputValidator.get_valid_options()
print(f"   Audiences: {valid_options['audiences']}")
print(f"   Tones: {valid_options['tones']}")

print("\n✅ Task 5 Completed Successfully!")
```

```python
class PromptGeneratorApp:
    """
    Complete Prompt Generator Application
    Combines validation, templates, and styles
    """

    def __init__(self):
        # Initialize validator
        self.validator = InputValidator()

        # Create style-based templates
        self.style_templates = {
            "teaching": PromptTemplate(
                input_variables=["topic", "audience", "tone"],
                template="Explain {topic} for {audience} in a {tone} teaching style with clear step-by-step instructions"
            ),
            "interview": PromptTemplate(
                input_variables=["topic", "audience", "tone"],
                template="Prepare {tone} interview questions about {topic} suitable for {audience} level candidates"
            ),
            "storytelling": PromptTemplate(
                input_variables=["topic", "audience", "tone"],
                template="Explain {topic} as a {tone} story for {audience}, making it engaging and memorable"
            )
        }

    def generate_prompt(self, topic: str, audience: str, tone: str, style: str,
                       validate: bool = True) -> Dict[str, any]:
        """
        Generate a complete prompt with validation

        Args:
            topic (str): Subject matter
            audience (str): Target audience
            tone (str): Communication tone
            style (str): Prompt style (teaching/interview/storytelling)
            validate (bool): Whether to validate inputs

        Returns:
            dict: Generated prompt and metadata
        """
        # Validate inputs if required
        if validate:
            validated = self.validator.validate_inputs(audience, tone, strict_mode=False)
            audience = validated["audience"]
            tone = validated["tone"]

        # Check if style is valid
        if style not in self.style_templates:
            raise ValueError(f"Invalid style '{style}'. Valid options: {list(self.style_templates.keys())}")

        # Generate prompt
        template = self.style_templates[style]
        prompt = template.format(topic=topic, audience=audience, tone=tone)

        return {
            "prompt": prompt,
            "metadata": {
                "topic": topic,
                "audience": audience,
                "tone": tone,
                "style": style
            }
        }

    def generate_all_styles(self, topic: str, audience: str, tone: str) -> Dict[str, str]:
        """
        Generate prompts for all available styles

        Args:
            topic (str): Subject matter
            audience (str): Target audience
            tone (str): Communication tone

        Returns:
            dict: Prompts for all styles
        """
        return {
            style: self.generate_prompt(topic, audience, tone, style)["prompt"]
            for style in self.style_templates.keys()
        }

    def get_available_styles(self) -> List[str]:
        """Get list of available prompt styles"""
        return list(self.style_templates.keys())

# Initialize the app
prompt_app = PromptGeneratorApp()

print("\n" + "=" * 60)
print("TASK 6: Prompt Generator App")
print("=" * 60)

# Example Test Cases
test_requests = [
    {
        "topic": "Neural Networks",
        "audience": "beginner",
        "tone": "fun",
        "style": "storytelling"
    },
    {
        "topic": "Quantum Computing",
        "audience": "expert",
        "tone": "formal",
        "style": "teaching"
    },
    {
        "topic": "Data Structures",
        "audience": "intermediate",
        "tone": "casual",
        "style": "interview"
    }
]

for i, request in enumerate(test_requests, 1):
    print(f"\n{'=' * 60}")
    print(f"📝 Request {i}")
    print(f"{'=' * 60}")

    result = prompt_app.generate_prompt(**request)

    print(f"\n📌 Metadata:")
    for key, value in result["metadata"].items():
        print(f"   {key.capitalize()}: {value}")

    print(f"\n✨ Generated Prompt:")
    print(f"   {result['prompt']}")

# Example: Generate all styles for one topic
print(f"\n{'=' * 60}")
print("🎨 All Styles for Single Topic")
print(f"{'=' * 60}")

all_styles = prompt_app.generate_all_styles(
    topic="Machine Learning",
    audience="beginner",
    tone="friendly"
)

for style, prompt in all_styles.items():
    print(f"\n🔸 {style.upper()}:")
    print(f"   {prompt}")

print("\n✅ Task 6 Completed Successfully!")
```

```python
class TemplateReusabilityTester:
    """
    Test template reusability with various inputs
    """

    def __init__(self):
        # Create a single reusable template
        self.universal_template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Create educational content about {topic} tailored for {audience} using a {tone} approach"
        )

    def test_reusability(self, test_cases: List[Dict]) -> List[Dict]:
        """
        Test the same template with multiple different inputs

        Args:
            test_cases (list): List of input combinations

        Returns:
            list: Results for each test case
        """
        results = []

        for i, test_case in enumerate(test_cases, 1):
            output = self.universal_template.format(**test_case)
            results.append({
                "test_number": i,
                "inputs": test_case,
                "output": output
            })

        return results

    def display_results(self, results: List[Dict]):
        """
        Display test results in formatted manner

        Args:
            results (list): Test results to display
        """
        for result in results:
            print(f"\n{'─' * 60}")
            print(f"🧪 Test #{result['test_number']}")
            print(f"{'─' * 60}")
            print(f"📥 Inputs:")
            for key, value in result['inputs'].items():
                print(f"   • {key}: {value}")
            print(f"\n📤 Output:")
            print(f"   {result['output']}")

# Initialize tester
reusability_tester = TemplateReusabilityTester()

# Create 5 diverse test cases
test_cases = [
    {
        "topic": "Artificial Intelligence",
        "audience": "beginner",
        "tone": "simple and friendly"
    },
    {
        "topic": "Blockchain Technology",
        "audience": "intermediate",
        "tone": "technical but accessible"
    },
    {
        "topic": "Quantum Mechanics",
        "audience": "expert",
        "tone": "formal and rigorous"
    },
    {
        "topic": "Web Development",
        "audience": "beginner",
        "tone": "fun and interactive"
    },
    {
        "topic": "Cybersecurity",
        "audience": "intermediate",
        "tone": "practical and actionable"
    }
]

print("\n" + "=" * 60)
print("TASK 7: Template Reusability Test")
print("=" * 60)
print("\n🎯 Objective: Use ONE template with 5 different inputs")
print(f"📋 Template Structure: {reusability_tester.universal_template.template}")

# Run reusability test
results = reusability_tester.test_reusability(test_cases)

# Display results
reusability_tester.display_results(results)

# Summary
print(f"\n{'=' * 60}")
print("📊 REUSABILITY TEST SUMMARY")
print(f"{'=' * 60}")
print(f"✅ Template Used: 1 (universal template)")
print(f"✅ Test Cases Run: {len(test_cases)}")
print(f"✅ Unique Outputs Generated: {len(set(r['output'] for r in results))}")
print(f"✅ Same Structure: YES")
print(f"✅ Different Outputs: YES")

print("\n✅ Task 7 Completed Successfully!")
```

```python
class CompletePipelineDemo:
    """
    Demonstrate the complete pipeline: Input → Validation → Template → Output
    """

    def __init__(self):
        self.validator = InputValidator()
        self.prompt_app = PromptGeneratorApp()

    def run_pipeline(self, topic: str, audience: str, tone: str, style: str):
        """
        Execute complete pipeline with all steps

        Args:
            topic (str): Subject matter
            audience (str): Target audience
            tone (str): Communication tone
            style (str): Prompt style
        """
        print("\n" + "🔄 " * 30)
        print("COMPLETE PIPELINE EXECUTION")
        print("🔄 " * 30)

        # Step 1: User Input
        print("\n📥 STEP 1: User Input")
        print(f"   Topic: {topic}")
        print(f"   Audience: {audience}")
        print(f"   Tone: {tone}")
        print(f"   Style: {style}")

        # Step 2: Validation
        print("\n✓ STEP 2: Input Validation")
        validated = self.validator.validate_inputs(audience, tone, strict_mode=False)
        print(f"   Validated Audience: {validated['audience']}")
        print(f"   Validated Tone: {validated['tone']}")

        # Step 3: Prompt Template Selection
        print("\n📝 STEP 3: Prompt Template Selection")
        print(f"   Selected Style: {style}")
        print(f"   Template Engine: Active")

        # Step 4: Dynamic Prompt Generation
        print("\n⚙️  STEP 4: Dynamic Prompt Generation")
        result = self.prompt_app.generate_prompt(
            topic=topic,
            audience=validated['audience'],
            tone=validated['tone'],
            style=style,
            validate=False  # Already validated
        )

        # Step 5: Output
        print("\n✨ STEP 5: Final Output")
        print(f"\n{'-' * 60}")
        print(result['prompt'])
        print(f"{'-' * 60}")

        return result

# Initialize pipeline
pipeline = CompletePipelineDemo()

# Run pipeline demo
pipeline.run_pipeline(
    topic="Large Language Models",
    audience="intermediate",
    tone="casual",
    style="teaching"
)

print("\n" + "=" * 60)
print("✅ COMPLETE PIPELINE DEMONSTRATION SUCCESSFUL")
print("=" * 60)
```
